```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add
```

```python
def folder(f_name): #this function creates a folder named "attacks" in the program directory.
    try:
        if not os.path.exists(f_name):
            os.makedirs(f_name)
    except OSError:
        print ("The folder could not be created!")
```

 # ---------------------------------------  { 1 }  -----------------------------------------

 ###### WLAN (WiFi) RSS database for fingerprinting positioning -> 1

```python
files_add=find_the_way("./data/1","rss.csv")
files_add
```

```python
for file in files_add :
    j_rssi = defaultdict(dict)
    j_coo = defaultdict(dict)
    rssis= pd.read_csv(file,header=None)
    cfile=file.replace("rss","coordinates")
    coos= pd.read_csv(cfile,header=None)
    coos=coos.values
    for ii,i in enumerate(rssis.values):
        for jj, j in enumerate(i):
            if j!=100:
                j_rssi[ii][jj]=j
        j_coo[ii]=[coos[ii][0],coos[ii][1]]
   
    with open(file.replace("csv","json"), 'w', encoding='utf-8') as f:
        json.dump(j_rssi, f, ensure_ascii=False, indent=4)

    with open(cfile.replace("csv","json"), 'w', encoding='utf-8') as f:
        json.dump(j_coo, f, ensure_ascii=False, indent=4)
```
