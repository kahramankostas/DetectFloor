```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add


```

```python
import csv
import random

def merge_csv_files(trainfiles, output_file, sample_ratio=1.0, has_header=True, seed=42):
    """
    Birden fazla CSV dosyasını bellek dostu şekilde birleştirir.
    
    Args:
        trainfiles (list): Birleştirilecek CSV dosyalarının yolları.
        output_file (str): Çıktı dosyasının yolu.
        sample_ratio (float): 0.0 - 1.0 arası örnekleme oranı. (1.0 = tüm satırlar)
        has_header (bool): Dosyalarda başlık satırı varsa True.
        seed (int): Rastgele örnekleme için seed değeri.
    """

    random.seed(seed)
    first_file = True

    with open(output_file, "w", newline="", encoding="utf-8") as fout:
        writer = None

        for file_path in trainfiles:
            print(f"Processing: {file_path}")

            with open(file_path, "r", encoding="utf-8") as fin:
                reader = csv.reader(fin)
                header = next(reader) if has_header else None

                # İlk dosyadan başlığı yaz
                if first_file:
                    if has_header and header:
                        writer = csv.writer(fout)
                        writer.writerow(header)
                    first_file = False

                # Satır satır oku
                for row in reader:
                    if sample_ratio < 1.0:
                        if random.random() > sample_ratio:
                            continue  # satırı atla
                    writer.writerow(row)

    print(f"✅ Merge completed: {output_file}")
    if sample_ratio < 1.0:
        print(f"✅ Sampling applied: {sample_ratio * 100:.1f}% of data kept.")
```
